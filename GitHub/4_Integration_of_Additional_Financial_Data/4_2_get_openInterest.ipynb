```python
import pandas as pd

# Assuming df is your DataFrame with 3510 rows and 535 columns
# Replace 'path/to/your/excel/file.xlsx' with the actual path to your Excel file.
file_path = '/content/drive/MyDrive/CUNY/Capstone/Data/market_price/ES_2010_2023_without_head.xlsx'
df = pd.read_excel(file_path, header=None)  # Assuming there is no header in the Excel file

# Assuming the first column is 'date', and the rest are your data columns
date_column = df.iloc[:, 0]  # Extract the 'date' column
data_columns = df.iloc[:, 1:]  # Extract the data columns

# Create a new DataFrame with the desired structure
result_df = pd.DataFrame(columns=range(df.shape[1]))

# Fill the date column in the new DataFrame
result_df.iloc[:, 0] = date_column

# Iterate over rows and shift non-null values to the right
for i in range(len(df)):
    non_null_values = data_columns.iloc[i].dropna().tolist()
    result_df.iloc[i, -len(non_null_values):] = non_null_values

# Display the resulting DataFrame
print(result_df)
```

```python
import pandas as pd

# Extract columns 0, 527, and 534
selected_columns_df = result_df.iloc[:, [0, 506, 513, 520, 527, 534]]

# Rename the columns
selected_columns_df.columns = ["Date", "far_month_openInterest_3", "far_month_openInterest_2", "far_month_openInterest_1", "far_month_openInterest_0", "near_month_openInterest"]

# Fill any potential NaN values in the DataFrame with 0
selected_columns_df = selected_columns_df.fillna(0)

# Create a new column "total_month_openInterest" by adding specific columns
selected_columns_df["total_month_openInterest"] = (
    selected_columns_df["far_month_openInterest_3"] +
    selected_columns_df["far_month_openInterest_2"] +
    selected_columns_df["far_month_openInterest_1"] +
    selected_columns_df["far_month_openInterest_0"] +
    selected_columns_df["near_month_openInterest"]
)

# Display the resulting DataFrame
print(selected_columns_df)

selected_columns_df.to_csv("selected_columns_df.csv", index=False)
```

```python
import sqlite3
import pandas as pd

def read_sqlite(conn_path, table_name):

    # Define your SQL query to retrieve data from a specific table
    query = "SELECT * FROM %s" %(table_name)
    with sqlite3.connect(conn_path) as conn:
        df = pd.read_sql_query(query, conn)

    return df

conn_path = '/content/drive/MyDrive/CUNY/Capstone/Data/2_wsj_titles/done/article_content/capstone.db'
table_name = 'e_mini_sp500_futuresHistorical_data_with_obtained_indexes_20100104_20231006'



df = read_sqlite(conn_path, table_name)

df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

print(df.head())
```

```python
from datetime import datetime, timezone
import pandas as pd
import time

def save_to_sqlite(df, conn_path, table_name):
    # Use a context manager to handle the connection
    with sqlite3.connect(conn_path) as conn:
        df.to_sql(table_name, conn, if_exists='replace', index=False)

    save_time = time.time()
    save_time_datetime = datetime.fromtimestamp(save_time, tz=timezone.utc)
    save_time_str = save_time_datetime.strftime('%Y-%m-%d %H:%M:%S')

    print("DataFrame has been saved on %s" % save_time_str)

conn_path = '/content/drive/MyDrive/CUNY/Capstone/Data/2_wsj_titles/done/article_content/capstone.db'
table_name = 'e_mini_sp500_futuresHistorical_data_with_obtained_indexes_20100104_20231006'

save_to_sqlite(merged_df, conn_path, table_name)
```
