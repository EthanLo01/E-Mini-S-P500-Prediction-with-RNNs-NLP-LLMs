# Fit Sentiment Analysis Weights to WSJ’s Article

```python
import pandas as pd

# Read the CSV file
feature_importance_df = pd.read_csv('/content/drive/MyDrive/CUNY/Capstone/Data/model/assign_label_weights_XGB/feature_importance.csv')

# Remove 'category_tags_' from the 'Feature' column
feature_importance_df['Feature'] = feature_importance_df['Feature'].str.replace('category_tags_', '')

# Convert 'Importance' to boolean based on the condition (e.g., > 0)
feature_importance_df = feature_importance_df[feature_importance_df['Importance'] > 0]

# Convert the DataFrame to a dictionary
feature_importance_dict = feature_importance_df.set_index('Feature')['Importance'].to_dict()

def calculate_weighted_sentiment(row, feature_importance_dict):
    tags = row['category_tags'].split()
    weights = [feature_importance_dict.get(tag, 0) for tag in tags]

    # Filter out non-numeric weights
    weights = [w for w in weights if isinstance(w, (int, float))]

    if weights:  # Check if weights list is not empty
        weight = sum(weights) / len(weights)

    else:
        weight = 0  # Assign a default weight if no tags or weights found

    # if weight!= 0:
    #     print(weight)
    row[['negative_weight', 'neutral_weight', 'positive_weight']] = row[['negative', 'neutral', 'positive']] * weight
    return row


WSJ_df = WSJ_df.apply(calculate_weighted_sentiment, axis=1, args=(feature_importance_dict,))

print(WSJ_df)
```

# Calculate Relevant Sentiment Analysis Data

```python
WSJ_df['SA_simple_score'] = WSJ_df['positive'] - WSJ_df['negative']
WSJ_df['SA_weighted_score'] = WSJ_df['positive_weight'] - WSJ_df['negative_weight']  # Corrected typo in 'negative_weight'
# WSJ_df['max_type'] = WSJ_df[['negative', 'neutral', 'positive']].idxmax(axis=1)
WSJ_df['Positive_count'] = (WSJ_df['positive'] != 0).astype(int)
WSJ_df['Neutral_count'] = (WSJ_df['neutral'] != 0).astype(int)
WSJ_df['Negative_count'] = (WSJ_df['negative'] != 0).astype(int)

WSJ_df['Positive_count_weighted'] = (WSJ_df['positive_weight'] != 0).astype(int)
WSJ_df['Neutral_count_weighted'] = (WSJ_df['neutral_weight'] != 0).astype(int)
WSJ_df['Negative_count_weighted'] = (WSJ_df['negative_weight'] != 0).astype(int)

print(WSJ_df.head())
```

# Group by S&P 500 Data and WSJ Sentiment Analysis Results

```python
# Initialize columns in df to zero
columns_to_set_zero = [
    # Sentiment columns
    'negative', 'neutral', 'positive',
    'negative_weight', 'neutral_weight', 'positive_weight',
    'SA_simple_score', 'SA_weighted_score',
    'Positive_count', 'Neutral_count', 'Negative_count',

    # Simple sentiment percentages and ratios
    'positive_score_percentage_simple', 'negative_score_percentage_simple',
    'positive_negative_score_ratio',

    # Weighted sentiment percentages and ratios
    'positive_score_percentage_weighted', 'negative_score_percentage_weighted',
    'positive_negative_score_ratio_weighted',

    # Count-based percentages and ratios
    'positive_count_percentage', 'negative_count_percentage',
    'positive_negative_count_percentage_ratio',

    # Weighted count-based percentages and ratios
    'Positive_count_weighted', 'Neutral_count_weighted', 'Negative_count_weighted',
    'positive_count_percentage_weighted', 'negative_count_percentage_weighted',
    'positive_negative_count_percentage_ratio_weighted'
]

# Set all the specified columns to zero in df
df[columns_to_set_zero] = 0

# Group by 'Business_Day' and sum up the respective columns
grouped = WSJ_df.groupby('Business_Day').sum()

# Compute simple sentiment percentages and ratios
total_sentiment = grouped[['negative', 'neutral', 'positive']].sum(axis=1)
grouped['positive_score_percentage_simple'] = grouped['positive'] / total_sentiment
grouped['negative_score_percentage_simple'] = grouped['negative'] / total_sentiment
grouped['positive_negative_score_ratio'] = grouped['positive'] / grouped['negative']

# Compute weighted sentiment percentages and ratios
total_weighted_sentiment = grouped[['negative_weight', 'neutral_weight', 'positive_weight']].sum(axis=1)
grouped['positive_score_percentage_weighted'] = grouped['positive_weight'] / total_weighted_sentiment
grouped['negative_score_percentage_weighted'] = grouped['negative_weight'] / total_weighted_sentiment
grouped['positive_negative_score_ratio_weighted'] = grouped['positive_weight'] / grouped['negative_weight']

# Compute count-based percentages and ratios
total_counts = grouped[['Positive_count', 'Neutral_count', 'Negative_count']].sum(axis=1)
grouped['positive_count_percentage'] = grouped['Positive_count'] / total_counts
grouped['negative_count_percentage'] = grouped['Negative_count'] / total_counts
grouped['positive_negative_count_percentage_ratio'] = grouped['Positive_count'] / grouped['Negative_count']

# Compute weighted count-based percentages and ratios
total_counts_weighted = grouped[['Positive_count_weighted', 'Neutral_count_weighted', 'Negative_count_weighted']].sum(axis=1)
grouped['positive_count_percentage_weighted'] = grouped['Positive_count_weighted'] / total_counts_weighted
grouped['negative_count_percentage_weighted'] = grouped['Negative_count_weighted'] / total_counts_weighted
grouped['positive_negative_count_percentage_ratio_weighted'] = grouped['Positive_count_weighted'] / grouped['Negative_count_weighted']

# Update df with the aggregated values
df.update(grouped[columns_to_set_zero].loc[df.index.intersection(grouped.index)])
```
