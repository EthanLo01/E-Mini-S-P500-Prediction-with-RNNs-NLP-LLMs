```python
import pandas as pd
import pandas_ta as ta



def fibo_retracement(df, window=5):
    high = df['High'].rolling(window=window).max()
    low = df['Low'].rolling(window=window).min()
    close = df['Close']

    fib_0_236 = close - (0.236 * (high - low))
    fib_0_382 = close - (0.382 * (high - low))
    fib_0_618 = close - (0.618 * (high - low))

    fibo_df = pd.DataFrame({
        'Fib_0.236': fib_0_236,
        'Fib_0.382': fib_0_382,
        'Fib_0.618': fib_0_618
    })

    df[f'Fib_0.236_{window}'] = fibo_df.iloc[:, 0]
    df[f'Fib_0.382_{window}'] = fibo_df.iloc[:, 1]
    df[f'Fib_0.618_{window}'] = fibo_df.iloc[:, 2]




def calculate_macd(df, fast=12, slow=26, signal=9):
    macd_result = ta.macd(df['Close'], fast=fast, slow=slow, signal=signal)
    df[f'macd_{fast}_{slow}_{signal}'] = macd_result[f'MACD_{fast}_{slow}_{signal}']
    df[f'macd_h_{fast}_{slow}_{signal}'] = macd_result[f'MACDh_{fast}_{slow}_{signal}']
    df[f'macd_s_{fast}_{slow}_{signal}'] = macd_result[f'MACDs_{fast}_{slow}_{signal}']


def calculate_kd(df, window=9, smooth_window=3):
    High, Low, Close = df['High'], df['Low'], df['Close']

    kd_df = ta.stoch(df['High'], df['Low'], df['Close'], window, smooth_window)

    df[f'STOCHk_{window}_{smooth_window}'] = kd_df.iloc[:, 0]
    df[f'STOCHd_{window}_{smooth_window}'] = kd_df.iloc[:, 1]

    return





# Calculate Moving Averages
df['SMA_5'] = ta.sma(df['Close'], length=5)
df['SMA_12'] = ta.sma(df['Close'], length=12)
df['SMA_26'] = ta.sma(df['Close'], length=26)
df['EMA_5'] = ta.ema(df['Close'], length=5)
df['EMA_12'] = ta.ema(df['Close'], length=12)
df['EMA_26'] = ta.ema(df['Close'], length=26)
df['SMA_50'] = ta.sma(df['Close'], length=50)
df['SMA_200'] = ta.sma(df['Close'], length=200)


# Calculate MACD
calculate_macd(df, fast=12, slow=26, signal=9)
calculate_macd(df, fast=5, slow=9, signal=3)
calculate_macd(df, fast=24, slow=52, signal=18)

# Calculate KD
calculate_kd(df, window=9, smooth_window=3)
calculate_kd(df, window=18, smooth_window=4)
calculate_kd(df, window=35, smooth_window=8)


# Calculate RSI
df['RSI_7'] = ta.rsi(df['Close'], length=7)
df['RSI_14'] = ta.rsi(df['Close'], length=14)



# Calculate Bollinger Bands
df[['BB_lower_band_5', 'BB_middle_band_5', 'BB_upper_band_5', 'BB_Width_5', 'BB_Percent_5']] = ta.bbands(df['Close'], length=5, std=2)
df[['BB_lower_band_10', 'BB_middle_band_10', 'BB_upper_band_10', 'BB_Width_10', 'BB_Percent_10']] = ta.bbands(df['Close'], length=10, std=2)
df[['BB_lower_band_20', 'BB_middle_band_20', 'BB_upper_band_20', 'BB_Width_20', 'BB_Percent_20']] = ta.bbands(df['Close'], length=20, std=2)


# Calculate Fibonacci Retracement
# You may need to define your own logic for Fibonacci retracement

# Calculate OBV
df['OBV'] = ta.obv(df['Close'], df['Vol.'])


# Calculate ADX
df[['ADX_14', 'DMP_14', 'DMN_14']] = ta.adx(df['High'], df['Low'], df['Close'], length=14)
df[['ADX_30', 'DMP_30', 'DMN_30']] = ta.adx(df['High'], df['Low'], df['Close'], length=30)
df[['ADX_50', 'DMP_50', 'DMN_50']] = ta.adx(df['High'], df['Low'], df['Close'], length=50)


# Calculate Fibo Retracement
fibo_retracement(df, window=5)
fibo_retracement(df, window=20)
fibo_retracement(df, window=100)


# Calculate Momentum Indicators (Short)
# df['MFI_7'] = ta.mfi(df['High'], df['Low'], df['Close'], df['Vol.'], length=7)
df['CCI_7'] = ta.cci(df['High'], df['Low'], df['Close'], length=7)
df['ROC_7'] = ta.roc(df['Close'], length=7)
df['WILLR_7'] = ta.willr(df['High'], df['Low'], df['Close'], length=7)

# Calculate Momentum Indicators (Long)
# df['MFI_14'] = ta.mfi(df['High'], df['Low'], df['Close'], df['Vol.'], length=14)
df['CCI_14'] = ta.cci(df['High'], df['Low'], df['Close'], length=14)
df['ROC_14'] = ta.roc(df['Close'], length=10)
df['WILLR_14'] = ta.willr(df['High'], df['Low'], df['Close'], length=14)






# Display the DataFrame with new columns
print(df)
```

```python
import sqlite3
from datetime import datetime, timezone
import pandas as pd
import time

def save_to_sqlite(df, conn_path, table_name):
    # Use a context manager to handle the connection
    with sqlite3.connect(conn_path) as conn:
        df.to_sql(table_name, conn, if_exists='replace')

    save_time = time.time()
    save_time_datetime = datetime.fromtimestamp(save_time, tz=timezone.utc)
    save_time_str = save_time_datetime.strftime('%Y-%m-%d %H:%M:%S')

    print("DataFrame has been saved on %s" % save_time_str)

conn_path = '/content/drive/MyDrive/CUNY/Capstone/Data/2_wsj_titles/done/article_content/capstone.db'
table_name = 'e_mini_sp500_futuresHistorical_data_with_obtained_indexes_20100104_20231006'

save_to_sqlite(df, conn_path, table_name)
```
