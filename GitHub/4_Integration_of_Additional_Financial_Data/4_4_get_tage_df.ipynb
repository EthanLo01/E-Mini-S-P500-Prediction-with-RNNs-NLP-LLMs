```python
import sqlite3
import pandas as pd

def read_sqlite(conn_path, table_name):

    # Define your SQL query to retrieve data from a specific table
    query = "SELECT * FROM %s" %(table_name)
    with sqlite3.connect(conn_path) as conn:
        df = pd.read_sql_query(query, conn)

    return df

conn_path = '/content/drive/MyDrive/CUNY/Capstone/Data/2_wsj_titles/done/article_content/capstone.db'
table_name = 'WSJ_articles_FinBERT'
WSJ_df = read_sqlite(conn_path, table_name)

selected_columns = ['Business_Day', 'category_tags', 'negative', 'neutral', 'positive']
WSJ_df = WSJ_df[selected_columns]

WSJ_df['Business_Day'] = pd.to_datetime(WSJ_df['Business_Day'])





print(WSJ_df.head())
```

```python
import pandas as pd

def expand_tags(df):
    new_rows = []
    for index, row in df.iterrows():
        tags = row['category_tags'].split(',')  # Split tags on commas
        if len(tags) > 1:
            # Expand rows for multi-tag entries
            for tag in tags:
                new_row = row.copy()
                new_row['category_tags'] = tag.strip().replace(" ", "")  # Remove leading/trailing spaces and extra whitespace
                new_rows.append(new_row)
        else:
            # Keep single-tag entries as-is
            new_rows.append(row)

    return pd.DataFrame(new_rows)

# Assuming you have your DataFrame WSJ_df
expanded_df = expand_tags(WSJ_df.copy(deep=True))  # deep=True preserves index
print(expanded_df)
```

```python
import pandas as pd

# Assuming you have the expanded DataFrame expanded_df

# Count unique values and their frequencies
tag_counts = expanded_df['category_tags'].value_counts()

# Filter the DataFrame, keeping only rows with tags that appear at least 100 times
expanded_df = expanded_df[expanded_df['category_tags'].isin(tag_counts[tag_counts >= 100].index)]

print(expanded_df)
```

```python
# Count the unique values in the 'category_tags' column
tag_counts = expanded_df['category_tags'].value_counts()

# Display the top 20 and bottom 20 values
top_20_tags = tag_counts.head(20)
bottom_20_tags = tag_counts.tail(20)

print("Top 20 Tags:")
print(top_20_tags)

print("\nBottom 20 Tags:")
print(bottom_20_tags)

# Print the number of unique tags
print("Tags Amt:", expanded_df['category_tags'].nunique())

```
