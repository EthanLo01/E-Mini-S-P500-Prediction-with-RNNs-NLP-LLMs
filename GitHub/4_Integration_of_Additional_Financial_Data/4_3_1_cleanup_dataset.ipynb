```python
import sqlite3
import pandas as pd

def read_sqlite(conn_path, table_name):

    # Define your SQL query to retrieve data from a specific table
    query = "SELECT * FROM %s" %(table_name)
    with sqlite3.connect(conn_path) as conn:
        df = pd.read_sql_query(query, conn)

    return df

conn_path = '/content/drive/MyDrive/CUNY/Capstone/Data/2_wsj_titles/done/article_content/capstone.db'
table_name = 'e_mini_sp500_futuresHistorical_data_with_obtained_indexes_20100104_20231006'



df = read_sqlite(conn_path, table_name)

df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')



print(df.head())
```

```python
df = df.sort_values(by='Date', ascending=True)
print(df.head())
```

```python
# Convert 'Price,' 'Open,' 'High,' and 'Low' columns to numeric
df[['Price', 'Open', 'High', 'Low']] = df[['Price', 'Open', 'High', 'Low']].replace('[\$,]', '', regex=True).astype(float)

# Rename the columns
df.rename(columns={'Price': 'Close'}, inplace=True)


# Display the updated DataFrame
print(df)
```

```python
import yfinance as yf

# Define the ticker symbol for E-mini S&P 500 futures (for example, ES=F)
ticker = "ES=F"

# Set the start and end dates for historical data
start_date = "2010-01-01"
end_date = "2023-12-14"

# Fetch historical data using yfinance
data = yf.download(ticker, start=start_date, end=end_date)

# Select the required columns
selected_columns = ["Open", "High", "Low", "Close", "Volume"]
selected_data = data[selected_columns]

# Calculate the daily percentage change
selected_data["Change %"] = selected_data["Close"].pct_change() * 100

# Display the result
print(selected_data)
```

```python
import pandas as pd

# Assuming df is your original DataFrame and selected_data is the DataFrame with updated values
# Make sure the "Date" column is in datetime format for proper indexing
df.index = pd.to_datetime(df.index)
selected_data.index = pd.to_datetime(selected_data.index)

# Update the "Vol." column in df with values from "Volume" in selected_data where dates match
df['Vol.'].update(selected_data['Volume'])
df['Change %'].update(selected_data['Change %'])


# Display the resulting DataFrame
print(df)
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Convert "Vol." column
def convert_volume(value):
    if value is not None and isinstance(value, str):
        if 'M' in value:
            return float(value.replace('M', '')) * 1000000
        elif 'K' in value:
            return float(value.replace('K', '')) * 1000
        else:
            return float(value)
    else:
        return value

df['Vol.'] = df['Vol.'].apply(convert_volume)

# Convert "Change %" column
def convert_change_percentage(value):
    if value is not None and isinstance(value, str) and '%' in value:
        return float(value.replace('%', '')) / 100
    else:
        return float(value)

df['Change %'] = df['Change %'].apply(convert_change_percentage)

# Display the resulting DataFrame
print(df)
```

```python
from datetime import datetime, timezone
import pandas as pd
import time

def save_to_sqlite(df, conn_path, table_name):
    # Use a context manager to handle the connection
    with sqlite3.connect(conn_path) as conn:
        df.to_sql(table_name, conn, if_exists='replace')

    save_time = time.time()
    save_time_datetime = datetime.fromtimestamp(save_time, tz=timezone.utc)
    save_time_str = save_time_datetime.strftime('%Y-%m-%d %H:%M:%S')

    print("DataFrame has been saved on %s" % save_time_str)

conn_path = '/content/drive/MyDrive/CUNY/Capstone/Data/2_wsj_titles/done/article_content/capstone.db'
table_name = 'e_mini_sp500_futuresHistorical_data_with_obtained_indexes_20100104_20231006'

save_to_sqlite(df, conn_path, table_name)
```
