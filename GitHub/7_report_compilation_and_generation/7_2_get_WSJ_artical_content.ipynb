```python
import sqlite3
import pandas as pd

def read_sqlite(conn_path, table_name):

    # Define your SQL query to retrieve data from a specific table
    query = "SELECT * FROM %s" %(table_name)
    with sqlite3.connect(conn_path) as conn:
        df = pd.read_sql_query(query, conn)

    return df

conn_path = '/content/drive/MyDrive/CUNY/Capstone/Data/2_wsj_titles/done/article_content/capstone.db'
table_name = 'WSJ_articles_FinBERT'
WSJ_df = read_sqlite(conn_path, table_name)

selected_columns = ['Business_Day', 'title', 'paragraphs', 'article','negative', 'neutral', 'positive']
selected_columns = ['Business_Day', 'article','negative', 'neutral', 'positive']
WSJ_df = WSJ_df[selected_columns]
WSJ_df[['negative_weight', 'neutral_weight' ,'positive_weight']] = 0

WSJ_df['Business_Day'] = pd.to_datetime(WSJ_df['Business_Day'])
WSJ_df.set_index('Business_Day', inplace=True)





print(WSJ_df.head())
```

```python
import pandas as pd
target_date = '2012-01-03'
WSJ_df.loc[target_date]
```

```python
positive_df = WSJ_df[WSJ_df['positive'] >0].loc[target_date]
positive_df
```

```python
negative_df = WSJ_df[WSJ_df['negative'] >0].loc[target_date]
negative_df
```

```python
from transformers import AutoTokenizer

def get_limited_text_content(text):
    # Replace 'bert-base-uncased' with the appropriate model name if needed
    tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

    # Count tokens
    token_count = len(tokenizer.encode(text, add_special_tokens=True))

    # Check if token count exceeds 4000
    if token_count > 4000:
        # Truncate text
        truncated_text = tokenizer.decode(tokenizer.encode(text, max_length=4000, truncation=True))
        print("Text truncated to 4000 tokens:", truncated_text)
        return truncated_text
    else:
        print("Number of tokens:", token_count)
        return text

positive_articles_text = '\n'.join(positive_df['article'].astype(str))

negative_articles_text = '\n'.join(negative_df['article'].astype(str))

positive_articles_text_result = get_limited_text_content(positive_articles_text)
negative_articles_text_result = get_limited_text_content(negative_articles_text)

print("Truncated Result:")
print(positive_articles_text_result)
print(negative_articles_text_result)
```
