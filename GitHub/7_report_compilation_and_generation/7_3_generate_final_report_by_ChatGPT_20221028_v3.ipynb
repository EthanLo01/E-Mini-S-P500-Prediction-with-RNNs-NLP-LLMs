```python
from transformers import AutoTokenizer

def get_limited_text_content(text):
    # Replace 'bert-base-uncased' with the appropriate model name if needed
    tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

    # Count tokens
    token_count = len(tokenizer.encode(text, add_special_tokens=True))

    # Check if token count exceeds 4000
    if token_count > 1500:
        # Truncate text
        truncated_text = tokenizer.decode(tokenizer.encode(text, max_length=1500, truncation=True))
        print("Text truncated to 1500 tokens:", truncated_text)
        return truncated_text
    else:
        print("Number of tokens:", token_count)
        return text

positive_articles_text = '\n'.join(positive_df['article'].astype(str))

negative_articles_text = '\n'.join(negative_df['article'].astype(str))

positive_articles_text_result = get_limited_text_content(positive_articles_text)
negative_articles_text_result = get_limited_text_content(negative_articles_text)

print("Truncated Result:")
print(positive_articles_text_result)
print(negative_articles_text_result)
```

```python
def create_dataset(dataset, lookback):
    X, y = [], []
    for i in range(len(dataset)-lookback):
        feature = dataset[i:i+lookback, :-1]
        target = dataset[i+1:i+lookback+1][-1][-1]
        X.append(feature)
        y.append(target)
    return torch.FloatTensor(X).to(device), torch.FloatTensor(y).view(-1, 1).to(device)

```
